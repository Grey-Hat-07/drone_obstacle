```python
import matplotlib.pyplot as plt
import numpy as np

# Starting coordinate for A
start_point = np.array([10, 10])

# Define the relative coordinates for the other vertices
relative_vertices = np.array([
    [0, 0],           # A
    [2, 0],           # B
    [3, np.sqrt(3)],  # C
    [2, 2*np.sqrt(3)],# D
    [0, 2*np.sqrt(3)],# E
    [-1, np.sqrt(3)]  # F
])

# Calculate absolute coordinates for all vertices
vertices = start_point + relative_vertices

# Close the hexagon by adding the first vertex at the end
vertices = np.vstack([vertices, vertices[0]])

# Points S, G, and W
S = [9, 11]
G = [13, 11]
W = [11, 8]

# Plot the hexagon
plt.figure(figsize=(8, 8))

# Plot vertices
plt.plot(vertices[:, 0], vertices[:, 1], marker='o', color='b')
plt.text(vertices[0][0], vertices[0][1], 'A', fontsize=12, ha='right')
plt.text(vertices[1][0], vertices[1][1], 'B', fontsize=12, ha='left')
plt.text(vertices[2][0], vertices[2][1], 'C', fontsize=12, ha='left')
plt.text(vertices[3][0], vertices[3][1], 'D', fontsize=12, ha='right')
plt.text(vertices[4][0], vertices[4][1], 'E', fontsize=12, ha='right')
plt.text(vertices[5][0], vertices[5][1], 'F', fontsize=12, ha='right')

# Plot points S, G, and W
plt.plot(S[0], S[1], marker='o', color='k')
plt.plot(G[0], G[1], marker='o', color='k')
plt.plot(W[0], W[1], marker='o', color='k')
plt.text(S[0], S[1], 'S', fontsize=12, ha='right')
plt.text(G[0], G[1], 'G', fontsize=12, ha='left')
plt.text(W[0], W[1], 'W', fontsize=12, ha='right')

# Connect S to W to G
plt.plot([S[0], W[0]], [S[1], W[1]], color='g')
plt.plot([W[0], G[0]], [W[1], G[1]], color='g')

plt.grid(True)
plt.xlim(8, 16)  # Adjusted the x-limit to make W visible
plt.ylim(7, 13)  # Adjusted the y-limit to make W visible
plt.title('Hexagon ABCDEF with Points S, G, and W')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
